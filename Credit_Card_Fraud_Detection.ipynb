```python
import pandas as pd

# Load the dataset
data = pd.read_csv("creditcard.csv")

# Display the first 5 rows
print(data.head())

# Check dataset structure
print(data.info())
```

```python
print(data.isnull().sum())  # Summarizes missing values
```

```python
data = data.dropna()  # Remove rows with missing values
```

```python
print(data.describe())  # Shows count, mean, std, min, max, etc.
```

```python
print(data.shape)  # Displays (rows, columns)
```

```python
print(data['Class'].value_counts())  # Count of 0s and 1s

# Visualize the distribution
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
data['Class'].value_counts().plot(kind='bar', color=['blue', 'orange'])
plt.title("Class Distribution (0 = Non-Fraud, 1 = Fraud)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
corr_matrix = data.corr()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.histplot(data['Amount'], bins=50, kde=True, color='blue')
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data['Time'], bins=50, kde=True, color='green')
plt.title("Transaction Time Distribution")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.show()
```

```python
fraud = data[data['Class'] == 1]
non_fraud = data[data['Class'] == 0]

sns.histplot(fraud['Amount'], bins=50, kde=True, color='red', label="Fraud")
sns.histplot(non_fraud['Amount'], bins=50, kde=True, color='blue', label="Non-Fraud")
plt.title("Amount Distribution (Fraud vs Non-Fraud)")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
sns.boxplot(x="Class", y="Amount", data=data)
plt.title("Boxplot of Amount by Class")
plt.show()
```

```python
# Check for outliers in 'Amount'
fraud_amounts = fraud['Amount']
q1 = fraud_amounts.quantile(0.25)
q3 = fraud_amounts.quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = fraud_amounts[(fraud_amounts < lower_bound) | (fraud_amounts > upper_bound)]
print(f"Outliers in fraud transactions: {len(outliers)}")
```
